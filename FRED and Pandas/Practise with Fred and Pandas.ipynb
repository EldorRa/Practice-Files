```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from fredapi import Fred

plt.style.use("fivethirtyeight")
pd.set_option("display.max_columns",500)
color_pal = plt.rcParams["axes.prop_cycle"].by_key()["color"]
```

```python
fred_key = "api_key"
```

```python
#Fred Object
fred = Fred(api_key = fred_key)
```

```python
#Searching the data about S&P and order by popularity column
sp_search = fred.search("S&P", order_by = "popularity")
sp_search.head()
```

```python
#Choosing series id data from dataset
sp500 = fred.get_series(series_id="SP500")
sp500.shape
```

```python
#Visualize the search data
sp500.plot(figsize=(15,5),title = "S&P 500");
```

```python
un_results = fred.search("unemployment")
un_results.head()
```
