Removing the outliers from dataset

```python
df = pd.read_csv("/kaggle/input/new-york-city-airbnb-open-data/AB_NYC_2019.csv")
df.head()
```

```python
df["price"].describe()
```

```python
print(df.price.mean())
print(df.price.median())
```

From the result we can see the percentile are lower than max but here max value is much higher so we have to remove the outlier to properly analyse the dataset


```python
df.price.quantile(1)
```

```python
df_99 = df.price.quantile(0.99)
df_99
```

```python
df[df.price>df_99]
```

```python
df.shape
```

```python
df_no_outliners = df[df.price<df_99]
df_no_outliners
```

```python
df_no_outliners.price.describe()
```
